## Anatomy of a Figure
<img style="float: right;" width="400" height="400" src="images/01_anatomy_of_plot.PNG">

Matplotlib graphs your data on <b>Figures</b> (e.g., windows, Jupyter widgets, etc.), each of which can contain one or more <b>Axes</b>, an area where points can be specified in terms of x-y coordinates (or theta-r in a polar plot, x-y-z in a 3D plot, etc.).

Below mentioned are the parts of Figure:
- Figure - The whole Figure.
- Axes - An <b>Axes</b> is an Artist attached to a Figure that contains a region for plotting data, and usually includes two (or three in the case of 3D) Axis objects (<b>be aware of the difference between Axes and Axis</b>) that provide ticks and tick labels to provide scales for the data in the Axes. Each Axes also has a title (set via <b>set_title()</b>), an x-label (set via <b>set_xlabel()</b>), and a y-label set via <b>set_ylabel()</b>)
- Artist - Basically, everything visible on the Figure is an Artist (even Figure, Axes, and Axis objects). This includes Text objects, Line2D objects, collections objects, Patch objects, etc. When the Figure is rendered, all of the Artists are drawn to the canvas. Most Artists are tied to an Axes; such an Artist cannot be shared by multiple Axes, or moved from one to another.

````python
# SYNTAX 2
# Initializing a figure (with 1 axes)
fig, ax = plt.subplots(figsize=(3, 3), facecolor='lightskyblue')
fig.suptitle("Figure Title")

# Adding Axes Title
ax.set_title("Axes Title", loc="left", fontsize="small")

# Adding Axes Grid
ax.grid(True)

# Adding Axes Color
ax.set_facecolor("yellow")

# Adding Annotations
ax.annotate("My Text", xytext=(0.2, 0.2), ha="center",
            xy=(0.5, 0.5), arrowprops={})

# Display the Figure
plt.show()
````

````python
# Initializing a Figure with 4 Axes (2*2 Grid)
fig, axs = plt.subplots(2, 2, figsize=(5, 5), 
                        facecolor='lightskyblue', 
                        layout='constrained')

# Set the color of axes to YELLOW
axs.set_facecolor('yellow')

plt.show()
````

````python
# Initializing a Figure with 4 Axes (2*2 Grid)
fig, axs = plt.subplots(2, 2, figsize=(5, 5), 
                        facecolor='lightskyblue', 
                        layout='constrained')

# Since axs is a numpy nd-array, we can iterate and make changes to each subplot
for ax in axs.ravel():
    ax.set_facecolor('yellow')

plt.show()
````

````python
# Initialize the figure with 4 axes (arranged in 2*2 Grid)
fig, axs = plt.subplots(2, 2, figsize=(5, 5), 
                   facecolor='lightskyblue', 
                   layout='constrained')
fig.suptitle('Figure')

title  = ['Axes_1', 'Axes_2', 'Axes_3', 'Axes_4']
color = ['yellow', 'violet', 'green', 'pink']

# Iterating over each axes
for idx, ax in enumerate(axs.ravel()):
    ax.set_facecolor(color[idx])
    ax.set_title(title[idx], loc='center', fontsize='medium')
    ax.set_xlabel("x Label")
    ax.set_ylabel("y Label")

plt.show()
````

## The implicit and the explicit interfaces

As noted above, there are essentially two ways to use Matplotlib:
- OO-style - Explicitly create Figures and Axes, and call methods on them (the "object-oriented (OO) style").
- pyplot-style - Rely on pyplot to implicitly create and manage the Figures and Axes, and use pyplot functions for plotting.

Matplotlib's documentation and examples use both the OO and the pyplot styles. In general, matplotlib suggest using the OO style, particularly for complicated plots, and functions and scripts that are intended to be reused as part of a larger project. However, the pyplot style can be very convenient for quick interactive work.

````python
plt.figure(figsize=(5, 2.7), layout='constrained')

plt.plot(data1, data1, label='linear')
plt.plot(data1, data1**2, label='quadratic')
plt.plot(data1, data1**3, label='cubic')

plt.title("Simple Plot")
plt.xlabel('x label')
plt.ylabel('y label')

plt.text(x=1.75, y=5.2, s="cubic", color='green')
plt.text(x=1.75, y=3.5, s="quadratic", color='orange')
plt.text(x=1.75, y=2, s="linear", color='blue')

plt.legend()

plt.show()
````

````python
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

ax.plot(data1, data1, label='linear', color='green')  # Plot some data on the axes
ax.plot(data1, data1**2, label='quadratic', color='orange')  # Plot more data on the axes...
ax.plot(data1, data1**3, label='cubic', color='blue')  # ... and some more

ax.set_title("Simple Plot")  # Add a title to the axes
ax.set_xlabel('x label')  # Add an x-label to the axes
ax.set_ylabel('y label')  # Add a y-label to the axes

ax.text(s="linear", x=1.75, y=2, color='green')
ax.text(s="quadratic", x=1.75, y=3.5, color='orange')
ax.text(s="cubic", x=1.75, y=5.2, color='blue')


ax.legend()  # Add a legend

plt.show()
````

````python
fig, axs = plt.subplots(1, 3, figsize=(10, 3), layout='constrained')
fig.suptitle("Comparision")

data = [data1, data1**2, data1**3]
labels = ['linear', 'quadratic', 'cubic']
colors = ['green', 'orange', 'blue']

for idx, ax in enumerate(axs.ravel()):
    ax.plot(data[0], data[idx], 
            label=labels[idx], 
            color=colors[idx])  # Plot some data on the axes
    
    ax.set_title("Simple Plot")  # Add a title to the axes
    ax.set_xlabel('x label')  # Add an x-label to the axes
    ax.set_ylabel('y label')  # Add a y-label to the axes
        
    ax.legend()  # Add a legend

plt.show()
````

### Univariate Non Visual Analysis

````python
def discrete_univariate_analysis(discrete_data):
    for col_name in discrete_data:
        print("*"*10, col_name, "*"*10)
        print(discrete_data[col_name].agg(['count', 'nunique', 'unique']))
        print('Value Counts: \n', discrete_data[col_name].value_counts())
        print()
````

````python
def numerical_univariate_analysis(numerical_data):
    for col_name in numerical_data:
        print("*"*10, col_name, "*"*10)
        print(numerical_data[col_name].agg(['count', 'min', 'max', 'mean', 'median', 'std']))
        print()
````

````python
discrete_df = df.select_dtypes(include=['object'])

numerical_df = df.select_dtypes(include=['int64', 'float64'])
````

````python
discrete_univariate_analysis(discrete_df)
````

````python
numerical_univariate_analysis(numerical_df)
````

### Univariate Visual Analysis

````python
# Histograms and Box Plots

fig, axs = plt.subplots(3, 2, figsize=(6, 8), layout="constrained")
fig.suptitle("Visual Univariate Analysis - Numerical Features")

cols = numerical_df.columns

for idx, ax in enumerate(axs):
    col_name = cols[idx]
    ax[0].hist(df[col_name])
    ax[0].set_title("Histogram Plot - " + col_name)
    ax[0].set_ylabel("Frequency")
    
    ax[1].boxplot(df[col_name])
    ax[1].set_title("Box Plot - " + col_name)    
````

````python
# Bar Plots

fig, axs = plt.subplots(2, 2, figsize=(6, 5), layout="constrained")
fig.suptitle("Visual Univariate Analysis - Categorical Features")

cols = discrete_df.columns

for idx, ax in enumerate(axs.ravel()):
    col_name = cols[idx]
    data = df[col_name].value_counts()
    ax.bar(data.index, data.values)
    ax.set_title("Bar Plot - " + col_name)
    ax.set_ylabel("Frequency")
````

````python
# Horizontal Bar Plots

fig, axs = plt.subplots(2, 2, figsize=(6, 5), layout="constrained")
fig.suptitle("Visual Univariate Analysis - Categorical Features")

cols = discrete_df.columns

for idx, ax in enumerate(axs.ravel()):
    col_name = cols[idx]
    data = df[col_name].value_counts()
    ax.barh(data.index, data.values)
    ax.set_title("Bar Plot - " + col_name)
    ax.set_ylabel("Frequency")
````
